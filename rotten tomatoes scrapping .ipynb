```python
import numpy as np
import csv
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url=requests.get('https://www.rottentomatoes.com/browse/movies_at_home/sort:critic_highest')
src=url.content
soup=BeautifulSoup(src,'lxml')
```

```python
movie_df=soup.find_all('div',{'class':'flex-container'})
```

```python
movie_names = []
stream_dates = []
audience_scores = []
audience_sentiments = []
```

```python
movie_df
```

```python
movie_name= movie_df[0].find('span',{'class':"p--small",'data-qa':"discovery-media-list-item-title"}).text.strip()
```

```python
stream_date=movie_df[0].find('span',{'class':"smaller",'data-qa':"discovery-media-list-item-start-date"}).text.strip()
```

```python
movie_rate=movie_df[0].find('score-pairs-deprecated')
```

```python
movie_rate
```

```python
rate=movie_rate.get('audiencescore')
audience_rate=movie_rate.get('audiencesentiment')
```

```python
import pandas as pd



# Loop through each movie in movie_df
for movie in movie_df:
    # Extract movie information
    name = movie.find('span', {'class': "p--small", 'data-qa': "discovery-media-list-item-title"}).text.strip().replace('#','')
    date = movie.find('span', {'class': "smaller", 'data-qa': "discovery-media-list-item-start-date"}).text.strip()
    rate_tag = movie.find('score-pairs-deprecated')
    audience_score = rate_tag.get('audiencescore') if rate_tag else None
    audience_sentiment = rate_tag.get('audiencesentiment') if rate_tag else None

    # Append extracted data to lists
    movie_names.append(name)
    stream_dates.append(date)
    audience_scores.append(audience_score)
    audience_sentiments.append(audience_sentiment)

# Create a DataFrame
data = {'Movie Name': movie_names, 'Stream Date': stream_dates, 'Audience Score': audience_scores, 'Audience Sentiment': audience_sentiments}
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```
