5. *revenue*: Numeric revenue from the sales, rounded to 2 decimal places.
   - Data Type: Float (for precision with two decimal places).
   - Data Validation:  i Ensured that the values are non-negative and rounded to two decimal places.
  i replaced the nulls with a average of each sales_method for a better results instead of dropping the nulls
  handling missing values by replacing them with average of each method

```python
df.dtypes
```

```python
df['revenue'].isnull().sum()
```

```python
df['revenue'].value_counts()
```

```python
avg_call = round(df[df['sales_method'] == 'Call'  ]['revenue'].mean(),2)
avg_Email = round(df[df['sales_method'] == 'Email'  ]['revenue'].mean(),2)
avg_Email_call = round(df[df['sales_method'] == 'Email + Call'  ]['revenue'].mean(),2)
df.loc[df['sales_method']=='Call', 'revenue'] = df.loc[df['sales_method']=='Call', 'revenue'].fillna(avg_call)
df.loc[df['sales_method']=='Email', 'revenue'] = df.loc[df['sales_method']=='Email', 'revenue'].fillna(avg_Email)
df.loc[df['sales_method']=='Email + Call', 'revenue'] = df.loc[df['sales_method']=='Email + Call', 'revenue'].fillna(avg_Email_call)
```
